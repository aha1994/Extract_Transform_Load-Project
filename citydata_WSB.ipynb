```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population'
df = pd.read_html(url)
cities_df = df[4]
cities = cities_df['City']
cities_df.head(2)
```

```python
#cleaning the city names and removing superscript
cities_clean = []
for x in cities:
    x = x.split('[')
    cities_clean.append(x)
cities_clean_df = pd.DataFrame(cities_clean)
cleaned_cities = cities_clean_df.drop([1], axis=1)
cleaned_cities_list = cleaned_cities[0]
cities_df['City Name'] = cleaned_cities_list
cities_df.head()
```

```python
cities = cities_df['City Name']
cities_ranks = cities_df['2018rank']
cities_pops = cities_df['2018estimate']
cities_states = cities_df['State[c]']


series = (cities,cities_ranks, cities_pops, cities_states)
cleaned_cities_df = pd.concat(series, axis=1)
cleaned_cities_df.head()

```

```python
url2 = 'https://en.wikipedia.org/wiki/List_of_metropolitan_statistical_areas'
metro_pop = pd.read_html(url2)
```

```python
metro_pop_DF = metro_pop[1]
metro_pop_DF.head()

#removing unicode characters from metro data column headers
metro_pop_DF.columns = [x.encode('utf-8').decode('ascii', 'ignore') for x in metro_pop_DF.columns]
metro_areas = metro_pop_DF['Metropolitanstatisticalarea']
metro_pop_DF.head(30)
```

```python
# Splitting metro area  names on the various delimiter characters
metro_cities = metro_pop_DF['Metropolitanstatisticalarea']
clean_metro_cities1 = []
for x in metro_cities:
    city = x.split('-')
    clean_metro_cities1.append(city[0])
```

```python
clean_metro_cities2 = []
for x in clean_metro_cities1:
    city = x.split('–')
    clean_metro_cities2.append(city[0])
```

```python
clean_metro_cities3 = []
for x in clean_metro_cities2:
    city = x.split(',')
    clean_metro_cities3.append(city[0])
```

```python
clean_metro_cities4 = []
for x in clean_metro_cities3:
    city = x.split('/')
    clean_metro_cities4.append(city[0])
```

```python
metro_pop_DF['Metro area name'] = clean_metro_cities4
metro_pop_DF.head()
```

```python
# Creating clean dataframes and renaming columns
metros = metro_pop_DF['Metro area name']
metro_ranks = metro_pop_DF['Rank']
metro_pops = metro_pop_DF['2018 estimate']

series2 = (metro_ranks, metros, metro_pops)
cleaned_metros_df = pd.concat(series2, axis=1)
cleaned_metros_df.head()
cleaned_metros_df_clean_names = cleaned_metros_df.rename(columns = {"Rank": "2018 Metro Rank", 
                                  "Metro area name":"Metro Name", 
                                  "2018 estimate": "2018 Metro Pop"}) 
cleaned_metros_df_clean_names.head()
```

```python
cleaned_cities_df.head()
cleaned_cities_df_clean_names = cleaned_cities_df.rename(columns = {"City Name": "City Name", 
                                  "2018rank":"2018 city rank", 
                                  "2018estimate": "2018 City Pop"}) 
cleaned_cities_df_clean_names.head()
```

```python
# dropped duplicate city names for both datasets, Columbus (Ohio and Georgia) etc
cleaned_metros_df_clean_names = cleaned_metros_df_clean_names.drop_duplicates('Metro Name', keep='first')
cleaned_cities_df_clean_names = cleaned_cities_df_clean_names.drop_duplicates('City Name', keep='first')

```

```python
# merge city and metro data on city name

merged_df = pd.merge(left=cleaned_cities_df_clean_names,right=cleaned_metros_df_clean_names, left_on='City Name', right_on='Metro Name')
density_score = (merged_df['2018 City Pop'])/(merged_df['2018 Metro Pop'])

merged_df['Density score'] = density_score

merged_df.head()
```
